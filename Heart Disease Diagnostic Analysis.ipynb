```python
def heart_disease(row):
    if row==0:
        return 'Absence'
    elif row==1:
        return 'Presence'
```

```python
Heart_data['Heart_Disease']=Heart_data['target'].apply(heart_disease)
```

```python
Heart_data.head()
```

```python
def sex(row):
    if row==0:
        return 'Female'
    elif row==1:
        return 'Male'
```

```python
Heart_data['sex1']=Heart_data['sex'].apply(sex)
```

```python
Heart_data.head()
```

```python
def age_range(row):
    if row>=29 and row<40:
        return 'Young Age'
    elif row>=40 and row<55:
        return 'Middle Age'
    elif row>55:
        return 'Elder Age'
```

```python
Heart_data['Age_Range']=Heart_data['age'].apply(age_range)
```

```python
def chest_pain(row):
    if row==1:
        return 'typical angina'
    elif row==2:
        return 'atypical angina'
    elif row==3:
        return 'non—anginal pain'
    else:
        return 'asymptotic'
```

```python
Heart_data['Chest_Pain']=Heart_data['cp'].apply(chest_pain)
```

```python
def slope(row):
    if row==0:
        return 'upsloping'
    elif row==1:
        return 'flat'
    elif row==2:
        return 'downsloping'
```

```python
Heart_data['slope1']=Heart_data['slope'].apply(slope)
```
